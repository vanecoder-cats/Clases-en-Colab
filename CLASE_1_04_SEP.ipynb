# Palabras reservadas

  Las palabras reservadas (keywords) corresponden a los nombres de las declaraciones que el interprete de Python incluye por defecto. No se debe utilizar dichas palabras para asignar nombres a dichos ojetos.

  El listado para dichas palabras reservadas puede consultando ingresando help("keyword") desde la interfaz.

Ejemplo:

```python
help("keywords")
```

# El espacio de nomres (namespace)
El espacio de nombres (namespace)contiene una relación de los onjetos existentes en la memoria del sistema y los nombre a los que esta ligado}



*  Python es un lenguaje de alto nivel en el que todos sus elementos son objetos incluyendo los tipos de datos basicos de tal forma que los terminos "valor" y "onjeto" son sinónimos.
*   Si un objeto no esta ligado al menos a un nombre, dicho objeto es deshechado por el interprete.
*  Un objeto puede tener mas de un nombre.
